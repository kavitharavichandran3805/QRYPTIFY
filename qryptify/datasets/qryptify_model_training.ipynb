```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.multioutput import MultiOutputClassifier
from xgboost import XGBClassifier
from sklearn.utils import shuffle
from sklearn.metrics import accuracy_score, f1_score, classification_report
import joblib
```

```python
train_df = pd.read_csv("features_dataset_train.csv")
test_df  = pd.read_csv("features_dataset_test.csv")
```

```python
train_df = shuffle(train_df, random_state=42).reset_index(drop=True)
test_df = shuffle(test_df, random_state=42).reset_index(drop=True)
```

```python
le_algo = LabelEncoder()
train_df['Algorithm_Label'] = le_algo.fit_transform(train_df['Algorithm'])
test_df['Algorithm_Label']  = le_algo.transform(test_df['Algorithm'])

le_type = LabelEncoder()
train_df['AlgorithmType_Label'] = le_type.fit_transform(train_df['Algorithm_Type'])
test_df['AlgorithmType_Label']  = le_type.transform(test_df['Algorithm_Type'])

le_cat = LabelEncoder()
train_df['Category_Label'] = le_cat.fit_transform(train_df['Category'])
test_df['Category_Label']  = le_cat.transform(test_df['Category'])
```

```python
X_train = train_df.drop(['Algorithm', 'Algorithm_Type', 'Category',
                         'Algorithm_Label', 'AlgorithmType_Label', 'Category_Label'], axis=1)
y_train = train_df[['Algorithm_Label', 'AlgorithmType_Label', 'Category_Label']]

X_test = test_df.drop(['Algorithm', 'Algorithm_Type', 'Category',
                       'Algorithm_Label', 'AlgorithmType_Label', 'Category_Label'], axis=1)
y_test = test_df[['Algorithm_Label', 'AlgorithmType_Label', 'Category_Label']]
```

```python
# Algorithm_Label model
model_algo = XGBClassifier(eval_metric='mlogloss', random_state=42, n_estimators=200)
model_algo.fit(X_train, y_train['Algorithm_Label'])

# AlgorithmType_Label model
model_type = XGBClassifier(eval_metric='mlogloss', random_state=42, n_estimators=200)
model_type.fit(X_train, y_train['AlgorithmType_Label'])

# Category_Label model
model_cat = XGBClassifier(eval_metric='mlogloss', random_state=42, n_estimators=200)
model_cat.fit(X_train, y_train['Category_Label'])
```

```python
y_pred_algo = model_algo.predict(X_test)
acc_algo = accuracy_score(y_test['Algorithm_Label'], y_pred_algo)
print(f"Accuracy (Algorithm_Label): {acc_algo:.4f}")
print(classification_report(y_test['Algorithm_Label'], y_pred_algo))

# Algorithm Type Prediction
y_pred_type = model_type.predict(X_test)
acc_type = accuracy_score(y_test['AlgorithmType_Label'], y_pred_type)
print(f"Accuracy (AlgorithmType_Label): {acc_type:.4f}")
print(classification_report(y_test['AlgorithmType_Label'], y_pred_type))

# Category Prediction
y_pred_cat = model_cat.predict(X_test)
acc_cat = accuracy_score(y_test['Category_Label'], y_pred_cat)
print(f"Accuracy (Category_Label): {acc_cat:.4f}")
print(classification_report(y_test['Category_Label'], y_pred_cat))
```
